## Importing and Cleaning the Data

```python
# Importing the dataset
import pandas as pd
flights = pd.read_csv('https://raw.githubusercontent.com/MarlaGoodman/CMS-Project/main/flights.csv')
```

```python
# Displaying the first 10 rows of data
flights.head(10)
```

```python
# Renaming the first column
flights.rename(columns = {'Unnamed: 0' : 'ID'}, inplace = True)
```

```python
# Displaying the first 10 rows
flights.head(10)
```

```python
# Adding year column
flights['year'] = 2017
flights.head(10)
```

## Stating Hypothesis and Further Data Manipulation

H0: No difference between average weekend delays and average weekday delays, or delays_weekend - delays_weekday = 0.

HA: Significant difference between average weekend delays and average weekday delays, or delays_weekend - delays_weekday != 0.

```python
# Creating a new date column
flights['Date'] = flights['year'].map(str) + '-' + flights['month'].map(str) + '-' + flights['day'].map(str)
flights.head(10)
```

```python
# Creating a new day of week column
flights['Date'] = pd.to_datetime(flights['Date'])
flights['day_of_week'] = flights['Date'].dt.day_name()
flights['day_of_week'].value_counts()
```

```python
# Creating a new total delays column
flights['total_delay'] = flights['dep_delay'] + flights['arr_delay']

# Fixing NaNs
total_delay_mean = flights['total_delay'].mean()
flights['total_delay'].fillna(value = total_delay_mean, inplace = True)
flights.head(10)
```

```python
# Creating samples
flights_weekend = flights[(flights.day_of_week == 'Saturday') | (flights.day_of_week == 'Sunday')]
flights_weekend_sample = flights_weekend.sample(190)
flights_weekday = flights[(flights.day_of_week != 'Saturday') & (flights.day_of_week != 'Sunday')]
flights_weekday_sample = flights_weekday.sample(190)
```

```python
flights_weekday_sample.head(10)
```

```python
flights_weekend_sample.head(10)
```

## Checking Conditions for T-Test

### Independence of Observations

```python
# Sample size of flights_weekday_sample
len(flights_weekday_sample) / len(flights)
```

```python
# Sample size of flights_weekend_sample
len(flights_weekend_sample) / len(flights)
```

Both simple random samples are less than 10% of the population, so they can be considered independent.

### Normally Distributed

```python
# Histogram of flights_weekday_sample
import matplotlib.pyplot as plt
flights_weekday_sample.hist(column = 'total_delay', bins = 25, grid = False)
```

```python
# Histogram of flights_weekend_sample
flights_weekend_sample.hist(column = 'total_delay', bins = 25, grid = False)
```

As we can see, both samples are definitely right skewed. However, this is understandable, as most flights' delays are only going to be a few minutes more than 0, with a few flights being excessively delayed. Additionally, with such a large sample size, this skewness should not bother us.

### Randomness

Both samples' data was taken as a simple random sample of the flights dataset.

## T Test for Independent Samples

For our t test, we will use a 95% confidence level (or an alpha level of 0.05).

```python
# Code found at: https://towardsdatascience.com/inferential-statistics-series-t-test-using-numpy-2718f8f9bf2f

# Importing the necessary libraries
import numpy as np
from scipy import stats
stats.ttest_ind(flights_weekday_sample['total_delay'], flights_weekend_sample['total_delay'])
```

## Results

Our p-value is 0.143, which is larger than our alpha level of 0.05. Therefore, we fail to reject the null hypothesis. We say that there is no statistically significant difference between the average weekend delays and the average weekday delays.

## Test 2: Weather Data

## Importing the Weather Data

```python
weather = pd.read_csv('https://raw.githubusercontent.com/MarlaGoodman/CMS-Project/main/weather_noaa.csv')
weather.head(10)
```

## Stating Hypothesis

H0: No difference between average snow induced delays and average rain induced delays, or delays_snow - delays_rain = 0.

HA: Significant difference between average snow induced delays and average rain induced delays, or delays_snow - delays_rain != 0.

## Data Cleaning

```python
# Breaking out the date variable into year, month, and day
weather['Date'] = pd.to_datetime(weather['Date'])
weather['day'] = weather['Date'].dt.day
weather['month'] = weather['Date'].dt.month
weather['year'] = weather['Date'].dt.year
weather.head(10)
```

```python
# Joining the data sets on year, month, and day
flights_weather = pd.merge(flights, weather,  how = 'inner', left_on=['year', 'month', 'day'], right_on = ['year', 'month', 'day'])
flights_weather.head(10)
```

```python
# Creating samples (Note: NOT including flights with both snow and rain)
flights_snow = flights_weather[(flights_weather.new_snow_bool == 1) & (flights_weather.rain_bool == 0)]
flights_snow_sample = flights_snow.sample(190)
flights_rain = flights_weather[(flights_weather.new_snow_bool == 0) & (flights_weather.rain_bool == 1)]
flights_rain_sample = flights_rain.sample(190)
```

```python
flights_snow_sample.head(10)
```

```python
flights_rain_sample.head(10)
```

## Checking Conditions for T Test

### Independence of Observations

```python
# Sample size of flights_snow_sample
len(flights_snow_sample) / len(flights_weather)
```

```python
# Sample size of flights_rain_sample
len(flights_rain_sample) / len(flights_weather)
```

Both simple random samples are less than 10% of the population, so they can be considered independent.

### Normally Distributed

```python
# Histogram of flights_snow_sample
flights_snow_sample.hist(column = 'total_delay', bins = 25, grid = False)
```

```python
# Histogram of flights_rain_sample
flights_rain_sample.hist(column = 'total_delay', bins = 25, grid = False)
```

As we can see, both samples are definitely right skewed. However, this is understandable, as most flights' delays are only going to be a few minutes more than 0, with a few flights being excessively delayed. Additionally, with such a large sample size, this skewness should not bother us.

It is interesting to note that the delays for flights with snow are slightly more centered around 0 (the distribution is slightly narrower), while the delays for flights with rain are slightly more spread out and have a few more extreme values / outliers. This could imply that flights with snow are slightly more likely to be delayed for a shorter amount of time.

### Randomness

Both samples' data was taken as a simple random sample of the flights_weather dataset.

## T Test for Independent Samples

For our t test, we will use a 95% confidence level (or an alpha level of 0.05).

```python
stats.ttest_ind(flights_snow_sample['total_delay'], flights_rain_sample['total_delay'])
```

## Results

Our p-value is 0.016, which is smaller than our alpha level of 0.05. Therefore, we reject the null hypothesis. We say that there is a statistically significant difference between the average snow delays and the average rain delays.

Based on the histograms we plotted above when checking our conditions, we could theorize that snow causes shorter average delays than rain does.
